```python
motos = pd.read_csv("../datasets/1.-Transportation/2017_cilindrada_motos.csv")

motos = motos.replace(to_replace = "No consta", value ="Unknown")
motos = motos.replace(to_replace = {"126 a 200":"126 a 200 cc", "Fins a 75 cc":"35 a 75 cc","Més de 750 cc":"750 a 1200 cc" })

motos_distr = pd.pivot_table(data=motos ,values="Nombre_motos", index = "Nom_Districte",columns="Cilindrada_motos",margins=True,aggfunc=sum
                            )

motos_distr
```

# 2017/2016 -1

```python
# First motor
growth_moto= (y_2017 / x_2016) -1
growth_moto
```

```python
# secondly accidents
growth_acc= (df_total_ac_2017 / df_total_ac_2016) -1
growth_acc
```

```python
# drop Desconegut and Unknown and inplace=True
growth_acc.drop(['Unknown','Desconegut'],inplace=True)
```

```python
growth_acc.rename(columns={"Neighborhood_Name":"Rate_accidents"},inplace=True)
growth_acc
```

```python
growth_moto
```

```python
# Make pivot table of motors
one_table= pd.pivot_table(growth_moto, values= 'Total_motors', columns= 'Cylinder',index=['Neighborhood_Name'])
one_table
```

```python
df_merged = one_table.merge(right=growth_acc, how="left",left_index=True,right_index=True)
df_merged
```

```python
df_merged.reset_index(inplace=True)
df_merged.head()
```
